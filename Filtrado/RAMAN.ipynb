```python
import pandas as pd
filename = '/content/M14 PT3 - Muestra 14 Punto 3 Modificado.csv'
data=pd.read_csv(filename,delimiter=',',decimal=",")
#pasar excel en formato numerico
data
```

```python
from numpy import *

x= data.X
y= data.Y
#pasar datos
x1= []
y1=[]
#print(x)
for c in range(len(x)):
    x1.append(float(x[c]))
    y1.append(float(y[c]))
#print(type(x1))
#print(x1)
#print(y1)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(26,10))
plt.title('Datos con ruido',fontsize = 22)
plt.xlabel('X',fontsize = 22)
plt.ylabel('Y',fontsize = 22)
plt.ylim(0,1800)
plt.xlim(200,3400)
plt.plot(x1,y1,color='gray')
plt.savefig('señalorginal.png')
```

```python
from numpy import *
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter as savgol

yf1 =  savgol(data.Y,11,0)

for u in range(len(y1)):
  data.Y[u] = yf1[u]
```

```python
data.to_csv('datafiltradaM14PT3.csv',sep=',')
```
